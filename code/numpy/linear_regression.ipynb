# Linear Regression using NumPy


## 1. Import all the necessary package will be used for computation .

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

print(np.__version__)
print(pd.__version__)
```

## 2. Read the input file using pandas library .

```python
data = pd.read_csv('HappinessAlcoholConsumption.csv')
data['alc'] = data['Beer_PerCapita'] + data["Spirit_PerCapita"] + data["Wine_PerCapita"]
df = data[['HappinessScore', 'alc']]
matrix = np.array(df.values, 'float')
matrix
```

```python
X = matrix[:,1]
y = matrix[:,0]
```

```python
# Feature Normalization

X = X/(np.max(X))
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y,'bo')
plt.ylabel('Happiness Score')
plt.xlabel('Alcohol consumption')
plt.legend(['Happiness Score'])
plt.title('Alcohol_Vs_Happiness')
plt.grid()
plt.show()
```

```python
def computecost(x,y,theta):
    a = 1/(2*m)
    b = np.sum(((x@theta)-y)**2)
    j = (a)*(b)
    return j
```

```python
#initialising parameter
m = np.size(y)
X = X.reshape([122,1])
x = np.hstack([np.ones_like(X),X])
theta = np.zeros([2,1])
print(theta,'\n',m)
```

```python
print(computecost(x,y,theta))
```

```python
def gradient(x,y,theta):
    
    alpha = 0.00001
    iteration = 2000

    #gradient descend algorithm
    J_history = np.zeros([iteration, 1]);
    
    for iter in range(0,2000):    
        error = (x @ theta) -y
        temp0 = theta[0] - ((alpha/m) * np.sum(error*x[:,0]))
        temp1 = theta[1] - ((alpha/m) * np.sum(error*x[:,1]))
        theta = np.array([temp0,temp1]).reshape(2,1)
        J_history[iter] = (1 / (2*m) ) * (np.sum(((x @ theta)-y)**2))   #compute J value for each iteration 
    return theta, J_history
```

```python
theta , J = gradient(x,y,theta)
print(theta)
```

```python
theta, J = gradient(x,y,theta)
J[1:3]
```

```python
from matplotlib import animation
from IPython.display import HTML


def init(): 
    line.set_data(X, x@theta)
    return line,

def animate(n):
    window = J[n:n+2]
    line.set_data(X,x@window)
    return line,

# blit=True re-draws only the parts that have changed.

fig, ax = plt.subplots(figsize=(5, 3))
plt.plot(X,y,'bo')
plt.ylabel('Happiness Score')
plt.xlabel('Alcohol consumption')
plt.legend(['HAPPY','LinearFit'])
plt.title('Alcohol_Vs_Happiness')

line, = ax.plot([], [], 'r', lw=2)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(J)-2, interval=200, blit=True)
anim.save('basic_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])

plt.show()



```

```python
#visualising J (theta0 , theta1)
theta0_vals = np.linspace(-5,10,100).reshape(1,100)
theta1_vals = np.linspace(-5,10,100).reshape(1,100)

#initialise J value to matrix of 0
J_vals = np.zeros([np.size(theta0_vals),np.size(theta1_vals)])

#fill J_vals
for i in range(0,np.size(theta0_vals)):
    for j in range(0,np.size(theta1_vals)):
        t = np.array([theta0_vals[:,i],theta1_vals[:,j]])
        J_vals[i,j] = computecost(x,y,t)
        
# Because of the way meshgrids work in the surf command, we need to
# transpose J_vals before calling surf, or else the axes will be flipped
J_vals = J_vals.T
```

```python
#surface plot for covergence
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure(figsize=[12.0,8.0])
ax = fig.add_subplot(111,projection ='3d')
ax.plot_surface(theta0_vals,theta1_vals,J_vals)
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J_vals')
plt.show()
```
